```python
fname = os.path.splitext(path_to_file)[0]
```

```python
tmp_py = fname + '_steric.py'
tmp_xyz = fname + '_transform.xyz'
os.remove(tmp_py)
os.remove(tmp_xyz)
```

```python
ccdata = cclib.io.ccread(path_to_file)
indexes = ccdata.atomnos
id_list = []
bv_list = []
for i in range(len(indexes)-20):
    index = i + 1
    mol = db.dbstep(path_to_file, atom1=index, b=True, quiet=True)
    bv = mol.bur_vol
    id_list.append(index)
    bv_list.append(bv)
df = pd.DataFrame({'atom_indexes': id_list, 'bv_list': bv_list})
```
